```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv('/content/Employee.csv')
print(df)
df.isnull().sum()
```

```python
df.isnull()
```

```python
d = {'Male': 0, 'Female': 1}
df['Gender'] = df['Gender'].map(d)
b = {'Bachelors': 0, 'Masters': 1}
df['Education'] = df['Education'].map(b)
df.fillna(0, inplace=True)
details = ['Education',	'Gender','Age','PaymentTier','ExperienceInCurrentDomain']
x = df[details]
y = df['LeaveOrNot']
```

```python
clf = svm.SVC()
clf.fit(x, y)
```

```python
y_pred = clf.predict(x)
accuracy = accuracy_score(y, y_pred)

print(f"Model Accuracy: {accuracy:.2f}")
```

```python

input_data = pd.DataFrame([[a, b, c, d, e]], columns=details)

prediction = clf.predict(input_data)
print('Leave Or Not: ', 'No Leave' if prediction == 0 else 'Leave')
print(f"Model Accuracy: {accuracy:.2f}")
```
