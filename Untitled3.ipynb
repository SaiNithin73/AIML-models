```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/Credit Card Customer Data.csv')
print(df.head(9))
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
X = df[['Total_Credit_Cards','Total_visits_bank','Total_visits_online','Total_calls_made']]
```

```python
Scalar = StandardScaler()
x_scalar = Scalar.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(x_scalar)
```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scalar)
centroids_pca = pca.transform(centroids)
```

```python
colours = ['red','blue','pink','green']
for i in range(len(x_pca)):
  plt.scatter(x_pca[i][0],x_pca[i][1],color=colours[labels[i]])

for i,c in enumerate(centroids_pca):
   plt.scatter(c[0], c[1], marker='x',s = 400, linewidths=3, color='brown')

plt.title('K-Means Clustering')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
df['cluster'] = labels
cluster_summary = df.groupby('cluster')[X.columns].mean()
print(cluster_summary)
```

```python
prediction = kmeans.predict([[6,2,7,0]])
print("total number of clusters:",prediction[0])
```
